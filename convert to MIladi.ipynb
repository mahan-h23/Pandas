```python
import pandas as pd
import numpy
import jdatetime
```

```python
df = pd.DataFrame({'date':["13990101","14010206","14000201","13980202"]})
```

```python
df
```

```python
for i in range (0, len(df['date'])):
    d = int(df.iloc[i,0][6:])
    m = int(df.iloc[i,0][4:6])
    y = int(df.iloc[i,0][0:4])
    df.iloc[i,0] = jdatetime.date(y,m,d).togregorian()
```
